```python
import matplotlib.pyplot as plt
import numpy as np
  
x = np.arange(-10, 11, 1)

def plot_graph(y,ylabel):
    plt.figure()
    plt.plot(x,y, 'o--')
    plt.grid(True)
    plt.xlabel("X")
    plt.ylabel(ylabel)
    plt.show()
```

```python
#STEP FUNCTION

y = list(map(lambda n: 1 if n>0.5 else 0, x))
plot_graph(y,"STEP(X)")
```

```python
#SIGMOID FUNCTION

y = 1 / (1 + np.exp(-x))
plot_graph(y, "Sigmoid(X)")
```

```python
#TANH FUNCTION

y = (np.exp(2*x) - 1) / (np.exp(2*x) + 1)
plot_graph(y, "TanH(X)")
```

```python
#RELU FUNCTION

y = list(map(lambda a: a if a>=0 else 0, x))
plot_graph(y,"ReLU(X)")
```

```python
#PRELU


alpha = 0.1

y = list(map(lambda a: a if a>=0 else a*0.1, x))
plot_graph(y,"ReLU(X)")
```

```python
#ERELU

import numpy as np

def eReLU(x, alpha=0.01):
    """ Exponential Rectified Linear Unit activation function """
    return np.where(x > 0, x, alpha*(np.exp(x) - 1))

x = np.linspace(-5, 5, 40)
y = list(map(lambda a: eReLU(a), x))

plot_graph(y, "eReLU(X)")
```
